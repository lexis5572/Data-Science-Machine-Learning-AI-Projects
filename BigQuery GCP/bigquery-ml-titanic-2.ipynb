```python
from google.cloud import bigquery

client = bigquery.Client(project='causal-hour-441123-k3')

query = """
CREATE MODEL `causal-hour-441123-k3.genai_marketing.ml4`
OPTIONS(
  model_type='BOOSTED_TREE_CLASSIFIER',
  data_split_method='AUTO_SPLIT',
  input_label_cols=['Survived']
) AS
SELECT
  PassengerId,
  Name,
  Pclass,
  Sex,
  SibSp,
  Parch,
  Fare,
  Embarked,
  Survived
FROM `causal-hour-441123-k3.genai_marketing.new-merged`;
"""

# Run the query
query_job = client.query(query)
query_job.result()  # Wait for the job to complete
print("Model created successfully.")
```

```python
from google.cloud import bigquery

client = bigquery.Client(project='causal-hour-441123-k3')

query = """
 SELECT
     *
   FROM
     ML.PREDICT(MODEL `causal-hour-441123-k3.new1.model2`,
                (SELECT * FROM `causal-hour-441123-k3.new2.test3`))
"""

# Run the query
query_job = client.query(query)
query_job.result()  # Wait for the job to complete
print("Model created successfully.")
```

```python
# Convert the results to a DataFrame

df = query_job.to_dataframe()

# Save DataFrame to CSV

df.to_csv('results1.csv', index=False)

print("Results saved to results.csv successfully.")
```
